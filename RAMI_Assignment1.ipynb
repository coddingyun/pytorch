```python
import matplotlib.pyplot as plt
from pydicom import dcmread

fpath = '/content/drive/MyDrive/1-072.dcm'
ds = dcmread(fpath)

print("Modality:",ds.Modality)
print("PatienttAge:",ds.PatientAge)
print("PatientSex:",ds.PatientSex)
print("PatientWeight:",ds.PatientWeight)
print()
data_array = ds.pixel_array
print("shape of pixel data:",data_array.shape)
print()
plt.figure(figsize=(10,10))
plt.imshow(data_array, cmap=plt.cm.gray)
plt.show()
```
